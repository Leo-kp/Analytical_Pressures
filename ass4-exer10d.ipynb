```python
import math
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
R=5
r=np.linspace(-5,5,100)
```

```python
v_laminar=1-(r/R)**2
```

```python
plt.plot(v_laminar,r)
plt.pause(1)
plt.title("profile $v(r)$_laminar")
plt.xlabel("$v_z$")
plt.ylabel("R")
```

```python
n=np.arange(1, 11, 1)

r_col = r[:, np.newaxis]
n_row = n[np.newaxis, :]
```

```python
v_turbulent=(1-abs(r_col)/R)**(1/n_row)
```

```python
for i in range(v_turbulent.shape[1]):
    plt.plot(v_turbulent[:, i], r, label=f'n={n[i]}')


plt.title("profile $v(r)$_turbulent")
plt.xlabel("$v_z$")
plt.ylabel("R")
plt.legend()
plt.pause(1)
```

```python
v_Stigler=1-(abs(r_col)/R)**(1/n_row)
for i in range(v_Stigler.shape[1]):
    plt.plot(v_Stigler[:, i], r, label=f'n={n[i]}')

plt.title("profile $v(r)$_Stigler_2014")
plt.xlabel("$v_z$")
plt.ylabel("R")
plt.legend()
plt.pause(1)
```

```python
u=1
p=np.linspace(-20,20,num=10)
y=np.linspace(0,1)

y_col= y[:, np.newaxis]
p_row= p[np.newaxis, :]

un=(u+p_row*(1-y_col))*y_col
```

```python
for i in range(un.shape[1]):
    plt.plot(un[:, i], y, label=f'p={round(p[i],2)}')

plt.title("profile $u(x,y)$")
plt.xlabel("$u$")
plt.ylabel("y")
plt.legend(bbox_to_anchor=(1.1, 1.05))
plt.pause(1)

```
